```python
import sys
sys.path.append("../../")

from src.sslHiC import get_reproducibility_score
from src.data.utils import cool_to_mats

import seaborn as sns
```

```python
resol = '500kb'
noise_levels = [0.05,0.1,0.15,0.2,0.25,0.3,0.4,0.5]
```

## DropEdge

```python
ref = cool_to_mats("dataset/noise/G401_replicate1_hg19_500kb.cool", ['chr10'])[0]

rep_scores = []
for noise_level in noise_levels:
    sim = cool_to_mats("dataset/noise/G401_replicate1_hg19_chr10_dropedge_22.cool", [f'chr10_{noise_level:.2f}'])[0]
    rep_score = get_reproducibility_score(ref, sim, resol)
    rep_scores.append(rep_score)
data = {'Noise levels':noise_levels, 'Reproducibility score': rep_scores}
sns.set()
sns.lineplot(data=data,x='Noise levels',y='Reproducibility score')
```
